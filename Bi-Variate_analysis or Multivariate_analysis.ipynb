## 1. Scatterplot (Numerical - Numerical)

**Shows the relationship b/w two variables i.e. how correlated() they are to eachother**

```python
sns.scatterplot(tips['total_bill'],tips['tip'])
plt.title('bi-variate analysis based on total-Bill and tip')
plt.show()
```

#### To Visualize the results based on the gender column , Multivariate analysis

```python
sns.scatterplot(tips['total_bill'],tips['tip'],hue=tips['sex'])
plt.title("Multivariate analysis based on tip vs (total_bill vs gender)")
plt.show()
```

#### Now we will put new parameter to gain more information based on style i.e. based on number of people they came top also depended on that

```python
sns.scatterplot(tips['total_bill'],tips['tip'],hue=tips['sex'],style=tips['size'])
plt.title('Multivariate Analysis on Tip vs (total_bill,gender,number-of-people)')
plt.show()
```

**Information from above, The groups that came bigger in number has given more tip**

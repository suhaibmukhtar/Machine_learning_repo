## Equation of Loss function of equation of curve of gradient (Forward Pass)

```python
def loss_function(m,b,data):
    slope,intercept=0,0
    loss=0
    predicted_y=0
    for i in range(len(x)):
        predicted_y=m*data['x'][i]
        loss+=(data['y'][i]-predicted_y)**2
    loss=loss/(2*len(data['x']))
    return loss
```

### Slope of gradient function (BackPropogation)

```python
def backpropogation_function(m,b,data,learning_rate):
    slope,intercept=0,0
    new_slope,new_intercept=0,0
    for i in range(len(data['x'])):
        predicted_y=m*data['x'][i]+b
        slope+=(data['y'][i]-predicted_y)*(-x[i]) #gradient of loss w.r.t slope
        intercept+=(data['y'][i]-predicted_y)*(-1) #gradient of loss w.r.t intercept
    slope=slope/len(data['x'])
    intercept=intercept/len(data['x'])
    m=m-learning_rate*(slope)
    b=b-learning_rate*(intercept)
    return float(m),float(b)
```

### How actually Linear Regression Finds the Best-Fit Line
